# Problem Statement

 ### Does studying more hours improve student marks?

### Mini Projecct 


```python
import pandas as pd

#  i Define a data Distonary that store some values

data = {
    'Student': ['A','B','C','D','E'],
    'Study_Hours': [1,2,3,4,5],
    'Marks': [45,55,65,75,85]
}

df = pd.DataFrame(data)
df
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['Study_Hours'], df['Marks'], color='green' )
plt.title('Study Hours vs Marks')
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.grid(True)
plt.show()
```

Insight (VERY IMPORTANT)

-  As study hours increase â†’ marks increase

- Strong positive relationship
