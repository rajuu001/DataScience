# Box Plot
- Q1 (1st quartile / 25th percentile)  the value below which 25% of the data falls.
- Q2 (2nd quartile / 50th percentile / median)  the value below which 50% of the data falls.
- Q3 (3rd quartile / 75th percentile)  the value below which 75% of the data falls.
- Q4 (4th quartile / 100th percentile / maximum)  the value below which 100% of the data falls.

```python
# Create a box plot for the 'total_bill' column
sns.boxplot(y=df["total_bill"])
plt.title("Box Plot of Total Bill")
plt.show()
```

## Plot multiple charts in one figure.
- The subplot() function lets you divide your plotting area into a grid and place multiple plots in different sections of that grid.

```python
# First subplot (1 row, 2 columns, first plot)

plt.figure(figsize=(14,10))

plt.subplot(2, 2, 1)
plt.plot([1, 2, 3], [4, 5, 6])
plt.title("Line Plot")

# Second subplot (1 row, 2 columns, second plot)
plt.subplot(2, 2, 2)
plt.bar([1, 2, 3], [4, 5, 6])
plt.title("Bar Chart")

# third
plt.subplot(2, 2, 3)
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [25, 35, 30, 10]

# plt.figure(figsize=(6,6))/
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, shadow=True)
plt.title('Fruit Distribution')

# third
plt.subplot(2, 2, 4)

# data
study_hours = [1, 2, 3, 4, 5]
exam_scores = [55, 60, 70, 75, 85]

# Create the scatter plot
plt.scatter(study_hours, exam_scores, color='green')

# Add labels and title
plt.title('Study Hours vs Exam Scores')
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')

# Show the plot
plt.grid(True)

plt.tight_layout()  # Prevents overlapping titles
plt.show()
```

```python
# Months of the year
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# Sales data for 2024 (in thousands)
sales_2024 = [50, 55, 53, 60, 65, 70, 75, 72, 68, 70, 74, 80]

# Sales data for 2025 (in thousands)
sales_2025 = [52, 58, 56, 62, 70, 75, 78, 76, 73, 77, 79, 85]

# Create the line plot
plt.plot(months, sales_2024, label='Sales 2024', color='blue', marker='o')
plt.plot(months, sales_2025, label='Sales 2025', color='green', marker='s')

# Add labels and title
plt.xlabel('Month')
plt.ylabel('Sales (in thousands)')
plt.title('Monthly Sales Trends: 2024 vs 2025')

# Show legend
plt.legend()

# Display the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Dummy market share data (in %)
brands = ['Apple', 'Samsung', 'Xiaomi', 'Oppo', 'Others']
market_share = [30, 25, 20, 15, 10]

plt.figure(figsize=(7,7))

plt.pie(
    market_share,
    labels=brands,
    autopct='%1.1f%%',
    startangle=140,
    shadow=True,
    explode=[0.1, 0, 0, 0, 0]  # explode Apple slice to highlight
)

plt.title('Smartphone Market Share 2025 (Dummy Data)')
plt.show()
```

```python
import plotly.express as px
import pandas as pd

# Sample Titanic data
data = {
    "Sex": ["male", "male", "female", "female"],
    "Status": ["Survived", "Dead", "Survived", "Dead"],
    "Count": [100, 500, 200, 80]
}

df = pd.DataFrame(data)

# bar chart
fig = px.bar(
    df,
    x="Sex",
    y="Count",
    color="Status",          # color by survived/dead
    barmode="group",         # group bars side by side
    title="Titanic Survival by Gender"
)

# Show chart
fig.show()
```
