### How many employees Resingned & Terminated in each Department ?

```python
r_emp = data[data["Status"]=="Resigned"].groupby("Department")["Status"].count()
r_emp
```

```python
T_emp = data[data["Status"]=="Terminated"].groupby("Department")["Status"].count()
T_emp
```

```python

r_emp.plot()
```

```python
plt.figure(figsize=(8,10))
r_emp.plot(x = r_emp.index,y = r_emp.values, kind = 'bar' ,color = "orange" ,label = "Resigned")
T_emp.plot(x = T_emp.index,y = T_emp.values ,kind = 'bar' ,color = "blue" ,label = "Terminated")
plt.legend(title= "Employee Status ")
plt.ylabel("Count of Employe ")
plt.grid()
plt.show()
```

```python
pf = data.groupby('Department')["Performance_Rating"].mean().astype(int)
plt.figure(figsize=(10,4))
pf.plot(x = pf.index ,y = pf.values ,kind = "bar" ,color = "lightgreen")
plt.title("Avg performance Rating department Wise ")
plt.ylabel("Ratings")
plt.show()
```
