#### Gender data ploat 

```python

ax = sns.countplot(x='Gender', data=df)

# Add counts on top of the bars
for bar in ax.containers:
    ax.bar_label(bar)

# Display the plot
plt.show()
```

#### Create bar ploat For Geneder 

```python
sales_gen = df.groupby(['Gender'], as_index = False )['Amount'].sum().sort_values(by = "Amount",ascending=False)

sns.barplot(x= 'Gender',y = 'Amount',data = sales_gen)
plt.show()
```

#### ploat the graf of sales by state 

```python
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

# --- 3. Plot the results ---
plt.figure(figsize=(12, 6)) # Optional: Adjusts the size of the plot

# Use the 'sales_state' DataFrame to plot directly
plt.bar(sales_state['State'], sales_state['Orders'])

# Adding titles and labels for clarity
plt.title('Top 10 States by Total Orders')
plt.xlabel('State')
plt.ylabel('Total Orders')
plt.xticks(rotation=45) # Optional
```
