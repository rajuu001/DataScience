```python
X = data.iloc[:,0:1]
```

```python
X
```

```python
y = data.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=2
)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
X_test.iloc[0]
```

```python
lr.predict(X_test.iloc[[0]])
```

```python
m = lr.coef_
b = lr.intercept_

```

```python
x = float(input("Enter you cgpa "))
y = m*x + b

print("Your pckage around : ",y)
```
